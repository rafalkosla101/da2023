```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats

import cmdstanpy
```

```python
cmdstanpy.install_cmdstan()
```

```python
F, L = len("Rafal"), len("Kosla")
N = (L + F) * 100
```

```python
model1 = cmdstanpy.CmdStanModel(stan_file='code_1.stan')
samples1 = model1.sample(data={'N': N}, fixed_param=True, iter_sampling=1, iter_warmup=0, chains=1)
df1 = samples1.draws_pd()
```

```python
model2 = cmdstanpy.CmdStanModel(stan_file='code_2.stan')
samples2 = model2.sample(data={'N': N}, fixed_param=True, iter_sampling=1, iter_warmup=0, chains=1)
df2 = samples2.draws_pd()
```

Excercise 2. Compare normal and student models for data from first file


```python
data_sim1 = {'N': N, 'y': df1.loc[:, 'y[1]':f'y[{N}]'].values.reshape((N))}
```

```python
model3 = cmdstanpy.CmdStanModel(stan_file='code_3.stan')
samples3 = model3.sample(data=data_sim1)
df3 = samples3.draws_pd()
```

```python
model4 = cmdstanpy.CmdStanModel(stan_file='code_4.stan')
samples4 = model4.sample(data=data_sim1)
df4 = samples4.draws_pd()
df4.head()
```

```python
loo = az.compare({"Normal model": az.from_cmdstanpy(samples3), "Student model": az.from_cmdstanpy(samples4)}, ic="loo", scale="deviance")
print(loo)

waic = az.compare({"Normal model": az.from_cmdstanpy(samples3), "Student model": az.from_cmdstanpy(samples4)}, ic="waic", scale="deviance")
print(waic)
```

```python
_, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
az.plot_compare(loo, ax=ax[0])
ax[0].set_title("LOO")
az.plot_compare(waic, ax=ax[1])
ax[1].set_title("WAIC")
plt.show()
```

Model based on student t distribution has generally lowerr loo and waic values and is simplier in a way.

Excercise 3. Compare models with different numbers of predictors


```python
data_sim21 = {'N': N, 'K': 1, 'y': df2.loc[:, 'y[1]': f'y[{N}]'].values.reshape((N)), 'X': df2.loc[:, 'X[1,1]': f'X[{N},1]'].values.reshape((N, 1))}
data_sim22 = {'N': N, 'K': 2, 'y': df2.loc[:, 'y[1]' : f'y[{N}]'].values.reshape((N)), 'X': df2.loc[:, 'X[1,1]' : f'X[{N},2]'].values.reshape((N, 2))}
data_sim23 = {'N': N, 'K': 3, 'y': df2.loc[:, 'y[1]' : f'y[{N}]'].values.reshape((N)), 'X': df2.loc[:, 'X[1,1]' : f'X[{N},3]'].values.reshape((N, 3))}
```

```python
model5 = cmdstanpy.CmdStanModel(stan_file='code_5.stan')
samples51 = model5.sample(data=data_sim21)
samples52 = model5.sample(data=data_sim22)
samples53 = model5.sample(data=data_sim23)
```

```python
loo = az.compare({"single value": az.from_cmdstanpy(samples51), "double value": az.from_cmdstanpy(samples52), "triple value": az.from_cmdstanpy(samples53)},
ic="loo", scale="deviance")
print(loo)
waic = az.compare({"single value": az.from_cmdstanpy(samples51), "double value": az.from_cmdstanpy(samples52), "triple value": az.from_cmdstanpy(samples53)},
ic="waic", scale="deviance")
print(waic)
```

```python
_, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
az.plot_compare(loo, ax=ax[0])
ax[0].set_title("LOO")
az.plot_compare(waic, ax=ax[1])
ax[1].set_title("WAIC")
plt.show()
```

waic and loo are teh best for single predictor which indicates that it is the best model
