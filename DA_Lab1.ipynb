```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import math
```

Ex 2.

```
uniform_distribution_list = np.random.uniform(0, 1, 1000)
normal_distribution01_list = np.random.normal(0, 1, 1000)
normal_distribution010_list = np.random.normal(0, 10, 1000)
beta_distribution11_list = np.random.beta(1, 1, 1000)
beta_distribution22_list = np.random.beta(2, 2, 1000)
beta_distribution39_list = np.random.beta(3, 9, 1000)
normal_multi_distribution1 = np.random.multivariate_normal([0,0], [[1,0],[0,1]], 10000)
normal_multi_distribution2 = np.random.multivariate_normal([0,0], [[2,1],[1,2]], 10000)
```

Ex 3.


```
count, bins, ignored = plt.hist(uniform_distribution_list, bins=40, density=True)
plt.plot(bins, np.ones_like(bins),color='r')
plt.title("Uniform distribution(0,1) histogram")
plt.show()
```

```
plt.hist(normal_distribution01_list, bins=30, density=True)
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title("Normal distribution(0,1) histogram")
plt.show()

```

```
plt.hist(normal_distribution010_list, bins=50, density=True)
mu = 0
variance = 10
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title("Normal distribution(0,10) histogram")
plt.show()
```

```
plt.hist(beta_distribution11_list, bins=40, density=True)
x = np.arange (0.01, 1, 0.01)
y = stats.beta.pdf(x,1,1)
plt.plot(x,y)
plt.title("Beta distribution(1,1) histogram")
plt.show()
```

```
plt.hist(beta_distribution22_list, bins=30, density=True)
x = np.arange (0.01, 1, 0.01)
y = stats.beta.pdf(x,2,2)
plt.plot(x,y)
plt.title("Beta distribution(2,2) histogram")
plt.show()
```

```
plt.hist(beta_distribution39_list, bins=40, density=True)
x = np.arange (0.01, 1, 0.01)
y = stats.beta.pdf(x,3,9)
plt.plot(x,y)
plt.title("Beta distribution(3,9) histogram")
plt.show()
```

```
alpha = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = np.abs(np.random.normal(0,1,1000)) 
mi = alpha + beta * x
y = np.random.normal(mi,sigma,1000)
```

Ex 6.

```
plt.axis([-6,6,-4,10])
plt.scatter(x,y)
for i in range(0,1000):
  y1 = alpha[i] + beta[i] * -6
  y2 = alpha[i]  + beta[i] * 6
  plt.plot([-6,6], [y1, y2], color='blue',alpha=0.1)
```
